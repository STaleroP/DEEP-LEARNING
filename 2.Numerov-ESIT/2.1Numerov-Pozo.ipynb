## 1. Marco Teórico: Mecánica Cuántica

### 1.1 La ecuación de Schrödinger independiente del tiempo

En mecánica cuántica, toda la información sobre el estado de una partícula está contenida en su **función de onda** $\Psi$. Para un sistema estacionario (energía bien definida), la dinámica queda descrita por la **ecuación de Schrödinger independiente del tiempo** (ESIT):

$$\hat{H}\,\psi(x) = E\,\psi(x)$$

donde el operador hamiltoniano en una dimensión es:

$$\hat{H} = -\frac{\hbar^2}{2m}\frac{d^2}{dx^2} + V(x)$$

Esto conduce a la ecuación diferencial ordinaria de segundo orden:

$$\boxed{-\frac{\hbar^2}{2m}\frac{d^2\psi}{dx^2} + V(x)\,\psi = E\,\psi}$$

### 1.2 Condiciones sobre la función de onda

Para que $\psi(x)$ represente un estado físico, debe satisfacer:

| Condición | Significado físico |
|:---|:---|
| $\psi$ continua y diferenciable | Densidad de probabilidad bien definida |
| $\psi \to 0$ en regiones clásicamente inaccesibles | La partícula no puede estar donde $V \to \infty$ |
| $\int_{-\infty}^{\infty} \lvert\psi\rvert^2\,dx = 1$ | Normalización: probabilidad total = 1 |

### 1.3 Interpretación probabilística

La cantidad $|\psi(x)|^2\,dx$ representa la probabilidad de encontrar la partícula en el intervalo $[x,\, x+dx]$. Los **eigenvalores** $E_n$ son los únicos valores de energía que produce soluciones físicamente aceptables — la **cuantización de energía**.

## 2. Pozo de Potencial Infinito — Solución Analítica

### 2.1 Definición del potencial

El **pozo de potencial infinito** (o *partícula en una caja*) de longitud $L$ se define como:

$$V(x) = \begin{cases} 0 & 0 < x < L \\ \infty & \text{en otro caso} \end{cases}$$

Las paredes infinitas imponen **condiciones de contorno de Dirichlet**:

$$\psi(0) = 0, \qquad \psi(L) = 0$$

### 2.2 Solución analítica

Dentro del pozo ($V = 0$), la ESIT se reduce a:

$$\frac{d^2\psi}{dx^2} = -k^2\,\psi, \qquad k^2 = \frac{2mE}{\hbar^2}$$

cuya solución general es $\psi(x) = A\sin(kx) + B\cos(kx)$. Aplicando $\psi(0)=0$ se tiene $B=0$; aplicando $\psi(L)=0$:

$$k_n L = n\pi \quad \Rightarrow \quad k_n = \frac{n\pi}{L}, \qquad n = 1, 2, 3, \ldots$$

Los **eigenvalores de energía** son:

$$\boxed{E_n = \frac{n^2\pi^2\hbar^2}{2mL^2}}$$

Las **eigenfunciones normalizadas** son:

$$\boxed{\psi_n(x) = \sqrt{\frac{2}{L}}\sin\left(\frac{n\pi x}{L}\right)}$$

### 2.3 Propiedades

- **Cuantización:** solo las energías $E_n \propto n^2$ son permitidas.
- **Energía de punto cero:** $E_1 > 0$, consecuencia del principio de incertidumbre de Heisenberg.
- **Paridad:** los estados tienen paridad alternante (simétricos/antisimétricos respecto al centro del pozo).
- **Ortogonalidad:** $\langle\psi_m|\psi_n\rangle = \delta_{mn}$.

## 3. El Método de Numerov — Derivación

### 3.1 Forma estándar de la ecuación

El método de Numerov es un integrador especializado para ecuaciones diferenciales de segundo orden **sin término de primera derivada**, de la forma:

$$\frac{d^2\psi}{dx^2} = f(x)\,\psi(x)$$

La ESIT se escribe exactamente en esta forma con:

$$f(x) = -k^2(x) = \frac{2m[V(x) - E]}{\hbar^2}$$

### 3.2 Derivación desde series de Taylor

Se discretiza el dominio en puntos $x_n = x_0 + nh$ con paso $h$. La expansión de Taylor de $\psi$ alrededor de $x_n$ es:

$$\psi_{n\pm 1} = \psi_n \pm h\psi'_n + \frac{h^2}{2}\psi''_n \pm \frac{h^3}{6}\psi'''_n + \frac{h^4}{24}\psi''''_n + \mathcal{O}(h^6)$$

Sumando $\psi_{n+1} + \psi_{n-1}$ se eliminan los términos de orden impar:

$$\psi_{n+1} + \psi_{n-1} = 2\psi_n + h^2\psi''_n + \frac{h^4}{12}\psi''''_n + \mathcal{O}(h^6) \tag{1}$$

La clave del método es aproximar $\psi''''_n$ usando diferencias finitas centradas sobre $\psi'' = f\psi$:

$$\psi''''_n \approx \frac{(f\psi)_{n+1} - 2(f\psi)_n + (f\psi)_{n-1}}{h^2} + \mathcal{O}(h^2)$$

Sustituyendo en $(1)$ y sustituyendo $\psi''_n = f_n\psi_n$:

$$\psi_{n+1} + \psi_{n-1} = 2\psi_n + h^2 f_n\psi_n + \frac{h^2}{12}\left[f_{n+1}\psi_{n+1} - 2f_n\psi_n + f_{n-1}\psi_{n-1}\right] + \mathcal{O}(h^6)$$

Reordenando para despejar $\psi_{n+1}$:

$$\boxed{\psi_{n+1} = \frac{2\left(1 - \dfrac{5h^2}{12}f_n\right)\psi_n\ -\ \left(1 + \dfrac{h^2}{12}f_{n-1}\right)\psi_{n-1}}{1 + \dfrac{h^2}{12}f_{n+1}} + \mathcal{O}(h^6)}$$

Para el pozo infinito, $f_n = -k_n^2 = -2mE/\hbar^2$, de modo que la fórmula queda:

$$\psi_{n+1} = \frac{2\left(1 + \dfrac{5h^2 k_n^2}{12}\right)\psi_n\ -\ \left(1 - \dfrac{h^2 k_{n-1}^2}{12}\right)\psi_{n-1}}{1 - \dfrac{h^2 k_{n+1}^2}{12}}$$

### 3.3 Ventajas del método

| Característica | Numerov | Diferencias finitas estándar |
|:---|:---:|:---:|
| Error local | $\mathcal{O}(h^6)$ | $\mathcal{O}(h^2)$ |
| Error global | $\mathcal{O}(h^4)$ | $\mathcal{O}(h^2)$ |
| Aplicabilidad | Solo $\psi'' = f(x)\psi$ | Ecuaciones generales |
| Costo computacional | Igual que diferencias finitas | — |

### 3.4 Condiciones iniciales y de frontera

Se necesitan **dos valores iniciales** para arrancar la recurrencia:
- $\psi_0 = 0$ (condición de frontera en $x=0$)
- $\psi_1 = h$ (valor infinitesimal arbitrario — la función se normaliza después)

La energía correcta $E_n$ se reconoce porque produce $\psi(L) \approx 0$. Para la solución analítica exacta se alimenta directamente $E_n$ al integrador.

## 4. Implementación Numérica

```python
# ── Standard library & scientific stack ──────────────────────────────────────
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact, IntSlider
from IPython.display import display

%matplotlib inline
plt.rcParams.update({'figure.dpi': 120, 'font.size': 11})
```

```python
# ── Physical constants (SI) ──────────────────────────────────────────────────
M_E  = 9.10938e-31   # electron mass          [kg]
HBAR = 1.05457e-34   # reduced Planck constant [J·s]

# ── Potential well parameters ────────────────────────────────────────────────
L    = 1.0e-9        # well width: 1 nm        [m]
N_PTS = 1000         # number of grid points

# ── Spatial grid ─────────────────────────────────────────────────────────────
x  = np.linspace(0.0, L, N_PTS)
dx = x[1] - x[0]

print(f"Grid: {N_PTS} points,  dx = {dx:.4e} m")
print(f"Well width L = {L:.2e} m")
```

### 4.1 Solución analítica de referencia

Las expresiones exactas permiten:
- **Validar** el integrador de Numerov.
- Cuantificar el **error numérico** como función del paso $h$.

$$E_n = \frac{n^2\pi^2\hbar^2}{2mL^2}, \qquad \psi_n(x) = \sqrt{\frac{2}{L}}\sin\!\left(\frac{n\pi x}{L}\right)$$

```python
def analytic_energy(n: int, L: float = L, m: float = M_E, hbar: float = HBAR) -> float:
    return (n**2 * np.pi**2 * hbar**2) / (2.0 * m * L**2)

def analytic_wavefunction(n: int, x: np.ndarray, L: float = L) -> np.ndarray:
    return np.sqrt(2.0 / L) * np.sin(n * np.pi * x / L)


# Display first 5 energy levels
print("Energías analíticas (primeros 5 niveles):")
print(f"  {'n':>3}   {'E_n [J]':>14}   {'E_n [eV]':>12}   {'E_n / E_1':>10}")
E1 = analytic_energy(1)
for n in range(1, 6):
    En = analytic_energy(n)
    print(f"  {n:>3}   {En:>14.6e}   {En/1.602e-19:>12.6f}   {En/E1:>10.4f}")
```

### 4.2 Integrador de Numerov

La implementación se divide en dos funciones con responsabilidades bien delimitadas:

- **`numerov_step`** — aplica la fórmula de recurrencia en un solo punto $x_{n+1}$.
- **`numerov_solve`** — recorre toda la malla y devuelve $\psi(x)$ normalizada.

```python
def numerov_step(
    psi_n: float,
    psi_nm1: float,
    k2_n: float,
    k2_nm1: float,
    k2_np1: float,
    h: float,
) -> float:
    """
    One Numerov recurrence step for ψ'' = -k²(x) ψ.

    Parameters
    ----------
    psi_n   : ψ at grid point n
    psi_nm1 : ψ at grid point n-1
    k2_n    : k²(x_n)
    k2_nm1  : k²(x_{n-1})
    k2_np1  : k²(x_{n+1})
    h       : grid spacing

    Returns
    -------
    ψ at grid point n+1

    Notes
    -----
    Implements the formula (error O(h^6)):

        ψ_{n+1} = [2(1 - 5h²k²_n/12) ψ_n  -  (1 + h²k²_{n-1}/12) ψ_{n-1}]
                  / (1 + h²k²_{n+1}/12)
    """
    h2 = h * h
    numerator   = (2.0 * (1.0 - (5.0 / 12.0) * h2 * k2_n) * psi_n
                   - (1.0 + (1.0 / 12.0) * h2 * k2_nm1) * psi_nm1)
    denominator =  1.0 + (1.0 / 12.0) * h2 * k2_np1
    return numerator / denominator
```

```python
def numerov_solve(
    E: float,
    x: np.ndarray,
    m: float = M_E,
    hbar: float = HBAR,
    normalize: bool = True,
) -> np.ndarray:
    """
    Solve the 1D Schrödinger equation for the infinite square well
    using the Numerov method.

    The well occupies x ∈ [x[0], x[-1]] with V = 0 inside.
    Boundary conditions: ψ(x[0]) = ψ(x[-1]) = 0.

    Parameters
    ----------
    E         : trial energy [J]
    x         : spatial grid [m]
    m         : particle mass [kg]
    hbar      : reduced Planck constant [J·s]
    normalize : if True, normalize ψ so that ∫|ψ|² dx = 1

    Returns
    -------
    psi : wave function evaluated on the grid
    """
    N  = len(x)
    h  = x[1] - x[0]

    # k²(x) = 2m(E - V)/ℏ²;  V = 0 inside the well
    k2 = (2.0 * m * E / hbar**2) * np.ones(N)

    psi = np.zeros(N)
    psi[0] = 0.0   # left boundary condition
    psi[1] = h     # seed — arbitrary small value, normalized later

    for i in range(1, N - 1):
        psi[i + 1] = numerov_step(
            psi[i], psi[i - 1],
            k2[i], k2[i - 1], k2[i + 1],
            h,
        )

    if normalize:
        norm = np.sqrt(np.trapezoid(psi**2, x))
        if norm > 0:
            psi /= norm

    return psi
```

### 4.3 Función de visualización

Se representa $\psi_n(x)$ y la densidad de probabilidad $|\psi_n(x)|^2$ junto con la geometría del pozo y las líneas de energía.

```python
def plot_wavefunction(n: int) -> None:
    """
    Plot the n-th eigenstate: wavefunction and probability density.

    Shows both the Numerov solution and the exact analytic result
    for comparison.

    Parameters
    ----------
    n : quantum number (positive integer)
    """
    E_n   = analytic_energy(n)
    psi_n = numerov_solve(E_n, x)
    psi_a = analytic_wavefunction(n, x)

    # Align signs: analytic ψ_n is positive near x=0 for odd n
    if np.sign(psi_n[N_PTS // 4]) != np.sign(psi_a[N_PTS // 4]):
        psi_n = -psi_n

    x_nm  = x * 1e9           # convert to nanometres for display
    prob   = psi_n**2
    prob_a = psi_a**2

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4.5))

    # ── Left panel: wave function ────────────────────────────────────────────
    ax1.plot(x_nm, psi_n, color='steelblue', lw=2,   label='Numerov')
    ax1.plot(x_nm, psi_a, color='coral',     lw=1.5, ls='--', label='Analítica')
    ax1.axvline(0,      color='k', lw=2.5, label='Pared del pozo')
    ax1.axvline(L*1e9,  color='k', lw=2.5)
    ax1.axhline(0,      color='gray', lw=0.7, ls=':')
    ax1.set_xlabel('$x$ [nm]')
    ax1.set_ylabel('$\\psi_n(x)$ [m$^{-1/2}$]')
    ax1.set_title(f'Función de onda  $n={n}$,  $E_n={E_n/1.602e-19:.4f}$ eV')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # ── Right panel: probability density ────────────────────────────────────
    ax2.fill_between(x_nm, prob, alpha=0.35, color='steelblue', label='Numerov')
    ax2.plot(x_nm, prob,   color='steelblue', lw=2)
    ax2.plot(x_nm, prob_a, color='coral', lw=1.5, ls='--', label='Analítica')
    ax2.axvline(0,      color='k', lw=2.5)
    ax2.axvline(L*1e9,  color='k', lw=2.5)
    ax2.set_xlabel('$x$ [nm]')
    ax2.set_ylabel('$|\\psi_n(x)|^2$ [m$^{-1}$]')
    ax2.set_title(f'Densidad de probabilidad  $n={n}$')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    # ── Diagnostics ──────────────────────────────────────────────────────────
    max_err = np.max(np.abs(psi_n - psi_a))
    norm    = np.trapezoid(psi_n**2, x)
    print(f"  Nivel n={n}:  E_n = {E_n:.6e} J  ({E_n/1.602e-19:.6f} eV)")
    print(f"  Error máximo |ψ_Numerov - ψ_analítica| = {max_err:.4e} m⁻¹/²")
    print(f"  Norma ∫|ψ|²dx = {norm:.8f}  (debe ser 1)")
```

## 6. Verificación y Análisis de Convergencia

### 6.1 Comparación de energías — Numerov vs. Analítica

Dado que usamos directamente $E_n$ analítica como entrada, el integrador devuelve la función de onda con error $\mathcal{O}(h^4)$ sobre la norma. Verificamos cuantitativamente:

- El **error en la norma** $|1 - \int|\psi_N|^2 dx|$ debe ser $\ll 1$.
- El **error máximo puntual** $\max|\psi_N - \psi_A|$ decrece como $h^4$ al refinar la malla.

```python
def print_energy_table(n_max: int = 8) -> None:
    """Print a comparison table of analytic vs. Numerov energy levels."""
    print(f"{'n':>3}  {'E_analítica [eV]':>18}  {'E_n/E_1':>8}  "
          f"{'Error norma':>12}  {'Error puntual máx.':>20}")
    print("-" * 72)
    for n in range(1, n_max + 1):
        E_n   = analytic_energy(n)
        psi_N = numerov_solve(E_n, x)
        psi_A = analytic_wavefunction(n, x)
        if np.sign(psi_N[N_PTS // 4]) != np.sign(psi_A[N_PTS // 4]):
            psi_N = -psi_N
        norm_err  = abs(1.0 - np.trapezoid(psi_N**2, x))
        point_err = np.max(np.abs(psi_N - psi_A))
        print(f"{n:>3}  {E_n/1.602e-19:>18.6f}  {n**2:>8.1f}  "
              f"{norm_err:>12.2e}  {point_err:>20.4e}")


print_energy_table()
```

### 6.2 Análisis de convergencia con el paso $h$

Se espera que el error del método de Numerov sea $\mathcal{O}(h^4)$: al reducir el paso a la mitad, el error debe disminuir por un factor $\sim 16$.

```python
def convergence_analysis(n: int = 1, n_pts_list=None) -> None:
    """
    Study how the Numerov error decreases as grid resolution increases.

    Parameters
    ----------
    n          : quantum level to analyse
    n_pts_list : list of grid sizes to test
    """
    if n_pts_list is None:
        n_pts_list = [50, 100, 200, 500, 1000, 2000]

    E_n = analytic_energy(n)
    steps, errors = [], []

    for N_pts in n_pts_list:
        xi   = np.linspace(0.0, L, N_pts)
        hi   = xi[1] - xi[0]
        psi_N = numerov_solve(E_n, xi)
        psi_A = analytic_wavefunction(n, xi)
        if np.sign(psi_N[N_pts // 4]) != np.sign(psi_A[N_pts // 4]):
            psi_N = -psi_N
        err = np.max(np.abs(psi_N - psi_A))
        steps.append(hi)
        errors.append(err)

    steps  = np.array(steps)
    errors = np.array(errors)

    # Fit slope in log-log space
    slope, intercept = np.polyfit(np.log(steps), np.log(errors), 1)

    fig, ax = plt.subplots(figsize=(7, 5))
    ax.loglog(steps, errors, 'o-', color='steelblue',
              label=f'Error máximo (pendiente ajustada = {slope:.2f})')
    ax.loglog(steps, np.exp(intercept) * steps**4,
              'k--', lw=1, label='Referencia $h^4$')
    ax.set_xlabel('Paso de malla $h$ [m]')
    ax.set_ylabel('Error máximo puntual $|\\psi_N - \\psi_A|$')
    ax.set_title(f'Convergencia del método de Numerov — nivel $n={n}$')
    ax.legend()
    ax.grid(True, which='both', alpha=0.3)
    plt.tight_layout()
    plt.show()

    print(f"  Pendiente ajustada en log-log: {slope:.3f}  (esperada ≈ 4.0)")


convergence_analysis(n=1)
```

### 6.3 Primeros niveles de energía — resumen gráfico

Se muestran los primeros $n_{\max}$ estados estacionarios superpuestos sobre el pozo de potencial, desplazados verticalmente por su energía — representación estándar en física cuántica.

```python
def plot_all_levels(n_max: int = 5) -> None:
    """
    Overlay the first n_max eigenstates on the potential well,
    each shifted vertically by its energy level.

    Parameters
    ----------
    n_max : number of levels to display
    """
    x_nm = x * 1e9
    colors = plt.cm.viridis(np.linspace(0.1, 0.9, n_max))

    fig, ax = plt.subplots(figsize=(9, 6))

    E_scale = analytic_energy(1) / 1.602e-19   # E_1 in eV (y-axis unit)

    for n, color in zip(range(1, n_max + 1), colors):
        E_n   = analytic_energy(n) / 1.602e-19  # eV
        psi   = numerov_solve(analytic_energy(n), x)
        # Scale wavefunction to fit between levels
        scale = 0.4 * E_scale
        ax.plot(x_nm, E_n + scale * psi,
                color=color, lw=1.8, label=f'$n={n}$,  $E_{n}={E_n:.3f}$ eV')
        ax.axhline(E_n, color=color, ls=':', lw=0.8, alpha=0.7)

    # Well walls
    E_top = analytic_energy(n_max) / 1.602e-19 * 1.3
    ax.axvline(0,      color='k', lw=3)
    ax.axvline(L*1e9,  color='k', lw=3)
    ax.axhline(0,      color='k', lw=1, ls='--', alpha=0.5)

    ax.set_xlim([-0.15, (L*1e9 + 0.15)])
    ax.set_ylim([-E_scale, E_top])
    ax.set_xlabel('$x$ [nm]')
    ax.set_ylabel('Energía [eV]')
    ax.set_title('Primeros niveles — Pozo de potencial infinito (Numerov)')
    ax.legend(loc='upper right', fontsize=9)
    ax.grid(True, alpha=0.2)
    plt.tight_layout()
    plt.show()


plot_all_levels(n_max=5)
```
